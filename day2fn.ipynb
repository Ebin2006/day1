```python
missing=df.isnull().sum()
ax=missing.plot(kind='bar',color='green')
plt.title("missing values")
plt.xlabel("Columns")
plt.ylabel(" Count of Missing Values")
plt.xticks(rotation=45)
for index,value in enumerate(missing):
    plt.text(index,value+5,str(value),ha='center',va='bottom',fontsize=10,fontweight='bold')
plt.show()
```

```python
df.drop(columns=['Cabin'])
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df.head()
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df.head()
```

```python
s=df['Survived'].value_counts()
print("Survived:\n",s)
```

```python
sns.countplot(x='Survived',data=df)
plt.title("survival count(0=Died,1=Survived)")
plt.xlabel("survived")
plt.ylabel("Passenger count")
plt.show()
```

```python
g=df['Sex'].value_counts()
print("Gender:\n",g)
```

```python
sg=df.groupby(['Sex','Survived']).size()
print("Survival by Gender:\n",sg)
```

```python
sns.countplot(x='Sex',hue='Survived',data=df,palette='Set1')
plt.title("Survivall by gender")
plt.xlabel("gender")
plt.ylabel("Passenger count")
plt.legend(title='survived')
plt.show()
```

```python
pc=df['Pclass'].value_counts()
print('passengeer class count:\n',pc)
```

```python
labels=['class1(upper)','class2(middle)','class3(lower)']
plt.figure(figsize=(7,7))
plt.pie(pc, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'seashell', 'olive'])
plt.title('Passenger Class Distribution (Pclass)', fontsize=14)
plt.axis('equal')
plt.show()
```

```python
sns.kdeplot(df['Age'],fill='True',color='skyblue')
plt.title('Age Distrubution(KDE Plot)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
sns.histplot(data=df,x='Age',hue='Survived',multiple='stack',palette='Set1')
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 40, 60, 100], labels=['Child', 'Teen', 'Adult', 'Middle Age', 'Senior'])
sns.countplot(x='AgeGroup', hue='Survived', data=df)
plt.title('Survival by Age Group')
plt.show()
```
